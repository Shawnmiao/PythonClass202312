```python
x = 10
y = 20
z = 30

print("Start")
if x == 10:
    print(" Nested If")
    if y == 20:
        print(" End of Nested If Block ")
    else:
        print(" End of Nested If-Else Block ")
elif y == 20:
    print(" Elif block ")
else:
    print(" Nested If")
    if z == 30:
        print(" End of Nested If Block ")
    else:
        print(" End of Nested If-Else Block ")
print("Stop")
```

#### （4）嵌套


```python
x=(int(input('please start values:')),int(input('please end values')))
x1=min(x)
x2=max(x)
nlist1=[]
for n in range(x1,x2+1):
    for i in range(2,n-1):
        print('the big loop n:',n)
        print('the small loop i:',i)
        if n%i==0:
            print('the big loop n:',n)
            print('the small loop i:',i)
            break
        
           
        
    nlist1.append(n)
print(nlist1)
```

```python
while True:
    account_info = {'miaomingyue':3456,'lihao':2345,'zhanghui':2234}        

    account_name = input("请输入您的用户名：")
    if account_name in account_info:
        password = input("请输入您的password")
        if account_info[account_name] ==int(password):
            print('进入系统')
            break
        else:
            print('您输入password错误，请重新输入')
    else:
        print('您输入的用户名不正确，请重新输入')
print('游戏开始')
```
