## 5.4参数传递 Passing arguments

```python
def test2(a,b=0,c=5):
    print(a,b,c)
test2(3,c=4,b=3)#b和c的顺序发生改变
```

```python
test2(3,4,6)
```

```python
test2(3)
```

一颗星参数：在定义函数时，如果参数名前加上一个“*”，则表示该参数就是一个可变长参数。在调用该函数时，如果依次序将所有的其他变量都赋值之后，剩下的参数将
会收集在一个元组中，元组的名称就是前面带星号的参数名。

两颗星参数：如果要收集不定量的关键字参数，可以在自定义函数时的参数前加“**”，这样多余的关键字参数就可以用字典方式被收集到变量中。

```python
def test_onestar(*a,b,c=2,d=4):
    print('a',a)
    print('b',b)
    print('c',c)
    print('d',d)
test_onestar(2,3,5,4,5,b=5)#此时位置参数须指定
```

```python
def test_twostar(b,c=2,d=6,**a):#双星参数放置在最后
    print(a)
    print(b)
    print(c)
    print(d)
test_twostar(5,4,x=5,y=6,z=7)
```

```python
def test_twostar(*b,c=2,d=6,**a):#双星参数放置在最后
    print(a)
    print(b)
    print(c)
    print(d)
test_twostar(1,2,3,5,4,x=5,y=6,z=7)
```

拆解元组：提供位置参数<br/>拆解字典：提供关键值参数<br/>

```python
def multi_test(x,y):
    print(x*y)
print('split tuple')
multi_test(*(4,5))
print('split dictionary')
multi_test(**{'x':5,'y':7})
```

python中的元素有可变和不可变之分，如整数、浮点数、字符串、元组等都属于不可变的，而列表和字典都属于可变的。在函数调用时，若提供的是不可变参数，那么在函数内部对其修改时，在函数外部其值是不变的；若提供是可变参数，则在函数内部对它修改时，在函数外部其值也会改变。

```python
def variables(value1,value2):
    value1=3
    value2[0]=0
    value2.append(8)
    print('in function',value1)
    print('in function',value2)
value1=9
value2=[5,6,7]
print('before function',value1)
print('before function',value2)
variables(value1,value2)
print('after function',value1)
print('after function',value2)
```

## 5.5变量的作用域

作用域可以分为：<br/>内置作用域：python预先定义的；<br/>
全局作用域：所编写的整个程序；<br/>局部作用域：某个函数内部范围

```python
def test_v():
    v=0
    print('within function{}'.format(v))
v=8#external
print('global:{}'.format(v))
test_v()
print('global:{}'.format(v))    
```

局部作用域可以引用全局作用域中的变量，但不能修改其在局部作用域之外的值；在局部作用域中给其赋值之前不得引用。

```python
t=6
def test_v2():
    #print(t)
    t=9
    print(t)
    
    
    
test_v2()
print(t)
```
