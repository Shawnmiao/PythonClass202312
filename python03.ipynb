### 3.3.3比较运算符

练习题：某小组的数学成绩如下：小王84分，小刘89分，小李100分，小周80分，小梅50分。请基于该小组数学成绩建立字典，并计算出该小组平均分。

```python
dic_math={"小王":84,'小刘':89,'小李':100,'小周':80,'小梅':50}
sv=sum(dic_math.values())
nv=len(dic_math)
av=sv/nv
av
```
