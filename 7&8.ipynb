# 7.异常与程序调试

异常即非正常状态，Python使用异常对象来表示异常，如果程序在编译或运行过程中发生错误，程序的执行过程就会发生变化，抛出异常对象。

程序进入异常处理，如果异常对象没有被处理或者捕捉，程序就会执行回溯（Traceback）来终止程序。

```python
printj
print(3)
```

```python
while True:  
    try:
        a = int(input("输入被除数："))
        b = int(input("输入除数："))
        c = a / b
        print("您输入的两个数相除的结果是：", c )
    except (ValueError, ArithmeticError):
        print("程序发生了数字格式异常、算术异常之一")
    except :
        print("未知异常")
    print("程序继续运行")
```

## 7.2手工异常

```python
def raiseTest():
    for i in range(7):
        if i==3:
            raise NameError
        print(i)
    print('end......')
raiseTest()
```

```python
def raiseTest2():
    for i in range(7):
        try:
            if i==3:
                raise NameError
        except NameError:
            print('Raise a NameError!')
        print(i)
    print('end......')
raiseTest2()
```
